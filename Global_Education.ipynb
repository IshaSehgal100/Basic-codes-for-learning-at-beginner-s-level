```python
plt.figure(figsize=(15,8))
plt.plot(data['Countries and areas'],data['Unemployment_Rate'])
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.xlabel('Countries')
plt.title('Country wise Unemployment Rates')
plt.show()
```

```python
#Top 10 countries with highest unemployement rates

top10_unemploy=data.groupby('Countries and areas')['Unemployment_Rate'].sum().reset_index().sort_values(by='Unemployment_Rate',ascending=True).tail(10)
top10_unemploy
```

```python
plt.bar(top10_unemploy['Countries and areas'],top10_unemploy['Unemployment_Rate'])
plt.xticks(rotation=90)
plt.ylabel('Unemployment_Rate')
plt.xlabel('Countries')
plt.title('Top 10 Countries with highest Unemployment Rate')
plt.show()
```

```python
#Top 50 countries with lowest unemployement rates

last50_unemploy=data.groupby('Countries and areas')['Unemployment_Rate'].sum().reset_index().sort_values(by='Unemployment_Rate',ascending=True).head(50)
last50_unemploy
```

These countries have zero unemployment rate

```python
plt.bar(last50_unemploy['Countries and areas'],last50_unemploy['Unemployment_Rate'],)
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.xlabel('Countries')
plt.title('Top 50 Countries with lowest Unemployment Rate')
plt.show()
```

```python
#top 50 female literacy rate countries
dfr=data[['Countries and areas','Youth_15_24_Literacy_Rate_Female']].sort_values(by='Youth_15_24_Literacy_Rate_Female',ascending=False).head(50)
dfr.head()
sns.set(rc={'figure.figsize':(11.7,8.27)})
g=sns.lineplot(x='Countries and areas',y='Youth_15_24_Literacy_Rate_Female',data=dfr);
g.set_title('Top 50 countries with highest Female Literacy Rates');
g.set_xticklabels(labels=dfr['Countries and areas'],rotation=90);
#plt.xticks(rotation=90)
```
